#1d newton method

```python
import math

def objective(x):
	return x**3 + 2

def derivative(x):
	return 3*x**2

def newton(x0 , e , mx) :
	xn = x0
	for n in range(0 , mx) :
		nfx = objective(xn)
		if abs(nfx) < e:
			return xn
		ndx = derivative(xn)
		xn = xn - (nfx/ndx)
		print(xn)
```

```python
ans = newton(-0.5 , 1e-10 , 10)
print(ans)
```

#1D newton method with mod


```python
def newton_mod(x0 , a , n , p) :
	i = 2
	cnt = 1
	xn = x0
	while (True) :
		xnn = xn * (2 - a * xn)
		if(abs(xnn) < (p ** i)) :
			xn = (p ** i) - abs(xnn)
			print("x" + str(cnt) + " is " + str(xn))
		else :
			k = math.ceil(abs(xnn) / (p ** i))
			xn = ((p ** i) * k) - abs(xnn)
			print("x" + str(cnt) + " is " + str(xn))
		if(i == n) : return xn
		i = i * 2
		cnt = cnt + 1
```

```python
ans = newton_mod(2 , 3 , 8 , 5)
print(str(ans) + " is congruence to 1 mod p^" + str(8))
```

```python
ans = newton_mod(3 , 5 , 32 , 7)
print(str(ans) + " is congruence to 1 mod p^" + str(32))
```

#2D newton method

```python
def objective1(X, Y):
    return 3*X**2 + 2*Y**2 - 25

def objective2(X, Y):
    return 2*X**2 - Y - 15
```

```python
from numpy import matrix
import numpy as np
count = 1
for x in range(0,100):
    if count == 1:
        X = 1
        Y = 1

    f1 = objective1(X , Y)
    f2 = objective2(X , Y)

    if f1 == 0 and f2==0:
        break
    print("\r\nIteration=%d" %count)
    count += 1
    print("f1= %f" %(f1))
    print("f2= %f" %(f2))

    A = 6*X
    B = 4*Y
    C = 4*X
    D = -1

    print(A,B)
    print(C,D)
    J = np.matrix([[A,B],[C,D]]).I  #Jacobian matrix
    f3 = np.matrix([[f1],[f2]])
    T = np.matrix([[X],[Y]])
    Delta = J*f3
    E = T-Delta
    X = np.squeeze(np.asarray(E[0]))
    Y = np.squeeze(np.asarray(E[1]))
    print(J)
    print("X=%f, Y=%f" %(X,Y))
```
